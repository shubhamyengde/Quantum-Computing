```python
from qiskit import *
```

```python
QR= QuantumRegister(4)
```

```python
CR= ClassicalRegister(4
                     )
```

```python
QC= QuantumCircuit(QR,CR)
```

```python
QC.h(QR[0])
QC.cx(QR[0],QR[1])
```

```python
QC.measure(QR,CR)
```

```python
QC.draw(output='mpl')
```

```python
simulator=Aer.get_backend('qasm_simulator')
```

```python
job=execute(QC,simulator,shots=1094)
```

```python
hist=job.result().get_counts()
```

```python
from qiskit.visualization import plot_histogram
```

```python
plot_histogram(hist)
```
